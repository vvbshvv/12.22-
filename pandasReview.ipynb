3. trainCG에서 Capital Gain을 타겟으로 LinearRegression 모델을 학습하고,
4. testCG에 있는 Capital Gain을 예측하여, 이 값으로 prepped 데이터 CG99999Idx 행들의 Capital Gain의 값을 대체한다.
