2개의 arr(a, b)간 연산에서 broadcasting 규칙
- a.shape = (i, j, k), b.shape = (l, m, n) 이라 하자.
- i, ㅣ는 0 axis의 size, 
- j, m은 1 axis의 size,
- k, n은 2 axis의 size.

broadcasting 조건:
1. 마지막 축의 axis의 size가 동일하거나, 어느 하나가 1이어야 한다.
 - k == n 또는 k == 1 또는 n == 1
2. 1의 조건이 만족될 때, 다음 마지막 축에서 동일한 조건이 성립되어야 한다.
 - j == m 또는 j == 1 또는 m == 1
3. 어느 한 쪽의 axis가 끝날 때가지 계속적으로 비교한다.
 - 어느 한 쪽의 axis가 끝났는 데 나머지 남아있으며, 그 만큼 복사가 이뤄진다.
 - l이 없었다면, b 쪽이 i 만큼 복사가 이뤄진다.

### 4) dot 연산의 이해
titanic.PNG

$$\begin{align}
y &= X \cdot w + b + e\\
&= u + b + e\\
&= \hat y + e
\end{align}$$
- y.shape = (n, ) : target. 예측하고자하는 값
- X.shape = (n, p) : Features. 설명변수.
- w.shape = (w, ) : 가중치. 예측값에 영향을 미치는 설명변수의 강도
- b는 상수 또는 b.shape = (n, ) : X이 0일때, y의 기본값.
- e.shape = (n, ) : 실제값과 예측값이 틀리는 정도
- $\hat y$.shape = (n, ): $\hat y = X \cdot w + b = u + b$. 예측값

dot 연산은 입력 피처(성, 나이, 운임)이 삶과 죽음에 영향을 미치는 :
- 방향(부호)과
- 강도(절대값)에 대한 가중치(w)에 대한 연산 결과를 나타낸다.


#### 코사인 법칙
![](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FdTLTo7%2FbtqLq2ZiLRV%2FELlfbaAOqklfOO1IKVdtn1%2Fimg.png)

위 그림에서 녹색으로 칠한 직각 삼각형에 주목하자. 직각 삼각형의 높이는 $b \sin \theta$이고, 밑변의 길이는  $a - b \cos \theta$이므로, 피타고라스의 정리에 의해서 다음 식이 성립한다.

$$\begin{align} c^2 &= (b \sin \theta )^2 + (a- b \cos \theta )^2 \\ \\ &= b^2 \sin^2 \theta + a^2 -2 ab \cos \theta + b^2 \cos^2 \theta \\ \\ &= a^2 + b^2 -2ab \cos \theta \end{align}$$


#### 두 벡터의 내적
두 벡터 $\mathbf{a}= \begin{bmatrix} a_1 & a_2 & \cdots & a_n \end{bmatrix}^T$ 
 와 $\mathbf{b}= \begin{bmatrix} b_1 & b_2 & \cdots & b_n \end{bmatrix}^T$ 가 있을 때, 두 벡터의 dot product 또는 내적(inner product)은 다음과 같이 정의한다.

 $$\mathbf{a} \cdot \mathbf{b} = a_1 b_1 + a_2 b_2 + \cdots + a_n b_n$$

내적은 위 식에서 보듯이 두 벡터의 성분을 차례로 곱한 다음에 모두 더하는 연산으로 정의한다. 보통 내적의 기호로 점(dot)을 이용하지만 다음과 같이 표기하기도 한다.

$$\mathbf{a} \cdot \mathbf{b} =(\mathbf{a},\ \mathbf{b}) = \mathbf{a}^T \mathbf{b}$$

내적 연산의 결과로 스칼라 값이 나오기 때문에 내적을 scalar product라고도 한다.정의에 의하면 내적은 다음과 같은 특성을 갖는다.

$$\begin{align} & \mathbf{a} \cdot \mathbf{b} = \mathbf{b} \cdot \mathbf{a} \\ \\ & \mathbf{a} \cdot (q_1 \mathbf{b} + q_2 \mathbf{c}) = q_1 ( \mathbf{a} \cdot \mathbf{b}) + q_2 ( \mathbf{a} \cdot \mathbf{c}) \\ \\ & \mathbf{0} \cdot \mathbf{a} = 0 \end{align}$$

여기서 $q_1$과 $q_2$는 임의의 상수다. 내적은 교환 법칙과 분배 법칙이 성립한다는 얘기로서 증명하는 것은 어렵지 않다. 한편 내적을 이용해서 벡터의 놈(norm)을 다음과 같이 정의한다.

$$\sqrt{ \mathbf{a} \cdot \mathbf{a} } = \| \mathbf{a} \|$$

내적에는 중요한 기하학적 의미가 있는데, 두 벡터 $a$와  $b$사의의 각(angle)$\theta$를 다음과 같이 내적을 이용해서 계산할 수 있다는 것이다.

$$\cos \theta = \frac{ \mathbf{a} \cdot \mathbf{b} } { \| \mathbf{a} \| \| \mathbf{b} \| }$$

#### 내적의 증명
여기서는 3차원 벡터에 대해서만 증명하지만 동일한 방법을 n-차원 벡터로도 확장시킬 수 있다. 원점 O에서 어떤 점 A와 B를 가리키는 벡터를 각각 a와 b라고 하자. 그러면 점 A에서 B까지의 벡터는 b-a 가 된다. 3개의 점은 한 평면을 이루므로 세 개의 벡터는 한 평면 상에 존재한다.

![](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FtrzV1%2FbtqLlA4gJwC%2Fnj5eegpFWUwXkKVfyQ3A6k%2Fimg.png)

두 벡터 a와 b의 사잇각을 $\theta$라고 하면, 코사인 법칙에 의해서 다음 식이 성립한다.

$$\| \mathbf{b}-\mathbf{a} \|^2 = \| \mathbf{a} \|^2 + \| \mathbf{b} \| ^2 - 2 \| \mathbf{a} \| \| \mathbf{b} \| \cos \theta$$

여기서
$$\begin{align} \| \mathbf{b}-\mathbf{a} \|^2 &= ( \mathbf{b}-\mathbf{a} ) \cdot ( \mathbf{b}-\mathbf{a} ) \\ \\ &= \mathbf{b} \cdot \mathbf{b} - 2 \mathbf{a} \cdot \mathbf{b} + \mathbf{a} \cdot \mathbf{a} \\ \\ &= \| \mathbf{b} \|^2 - 2 \mathbf{a} \cdot \mathbf{b} + \| \mathbf{a} \| ^2 \end{align}$$

이므로
$$\mathbf{a} \cdot \mathbf{b} = \| \mathbf{a} \| \| \mathbf{b} \| \cos \theta$$

가 성립한다.

#### 내적의 성질
만약 두 벡터의 내적이 0이면 두 벡터는 서로 직각이다.

- w.shape = (p,) : 데이터 X의 각 변수들이 y를 예측하는 데, 기여하는 가중치
- b.shape = (1,) : bias. 데이터 X가 모두 0일 때의 예측 값.

X = (1, X) 이고, w = (b, w) = (w0, w)로 변경하면:
$$\begin{align}
y &= X \cdot w + \epsilon \\
&= \hat y + \epsilon
\end{align}$$

e가 최소화되길 원한다:
$$\begin{align}
e &= y - \hat y \\
&= y - X \cdot w + \epsilon 
\end{align}$$


e가 최소화되길 원한다:
$$\begin{align}
e &= y - \hat y \\
&= y - X \cdot w \\
||\epsilon||^2 &= \epsilon \cdot \epsilon \\
&= (y - X \cdot w)^T \cdot (y - X \cdot w) \\
&= y^T \cdot y - w^T \cdot X^T \cdot y - y^T \cdot X \cdot w + w^T \cdot X^T \cdot X \cdot w \\
\cfrac {d ||\epsilon||^2}{d w} &= - 2 X^T \cdot y + 2 X^T \cdot X \cdot w\\
\cfrac {d ||\epsilon||^2}{d w} (w = \hat w) &= - 2 X^T \cdot y + 2 X^T \cdot X \cdot w = 0 \\
\hat w &= (X^T \cdot X)^{-1} \cdot X^T \cdot y
\end{align}$$

참고로:
$(X \cdot w)^T == w^T \cdot X^T$

### 2) 머신러닝 방법
$$\cfrac {d ||\epsilon||^2}{d w} = - 2 X^T \cdot y + 2 X^T \cdot X \cdot w$$

우리는 n개의 학습 데이터를 가지고 있다. 
- batch개 만큼의 데이터로 단계적인 학습을 진행합니다. 
- n / batch 번 학습을 하면, 전체 데이터를 1번 학습하는 것이다.
- 전체 데이터를 Epoch라는 숫자만큼 학습합니다.
- 매 학습마다 학습 파라미터를 아래와 같이 업데이트를 합니다.
$$w(t+1) = w(t) - \alpha \cfrac {d ||\epsilon||^2}{d w}$$
- $\alpha$ : learning rate (~ $10^{-4}$)

````python
n, p = Xp.shape
n, p
````

````python
np.random.seed(123)
w = np.random.random(p) - .5
````

````python
EPOCH = 100000
BATCH = 30
a = 10**(-4)
Losses = []

randIdx = np.arange(n)
for e in range(EPOCH):
    np.random.shuffle(randIdx)
    Loss = 0
    for startIdx in range(0, n, BATCH):
        batchIdx = randIdx[startIdx:startIdx+BATCH]
        X_batch, y_batch = Xp[batchIdx], y[batchIdx]
        yhat_batch = X_batch.dot(w)
        Loss += (y_batch - yhat_batch).dot(y_batch - yhat_batch)
        dw = X_batch.T.dot(X_batch).dot(w) - X_batch.T.dot(y_batch)
        w = w - a * dw
    Losses.append(Loss/n)
w
````

````python
lim.intercept_, lim.coef_
````

$$\cfrac {d ||\epsilon||^2}{d w} = - 2 X^T \cdot y + 2 X^T \cdot X \cdot w$$
$$w(t+1) = w(t) - \alpha \cfrac {d ||\epsilon||^2}{d w}$$
